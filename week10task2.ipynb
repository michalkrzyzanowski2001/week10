```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
def gauss(x, sigma=1, mu=3):
    return stats.norm.pdf(x, mu, sigma)
x = np.linspace(-2,8,10000)
plt.figure(figsize=(12, 8))
plt.plot(x, gauss(x, sigma=1), label='Normal distribution with 10000')
plt.legend(loc='upper left')
plt.show()
```

```python
std100 = norm.rvs(loc=3,size=100)
std10000 = norm.rvs(loc=3,size=10000)
plt.figure(figsize=(12, 8))
plt.hist(std100,bins=10,density=True,alpha=0.5,label="Normal Distribution with 100")
plt.hist(std10000,bins=100,density=True,alpha=0.5,label="Normal Distribution with 10000")
plt.plot(x, gauss(x, sigma=1), label='Normal distribution',color="red")
plt.axvline(x=3,color="Black",label="Mean")
plt.legend()
plt.xticks(ticks=[0,1,2,3,4,5,6],labels=["3","2","","0","","2","3"])
plt.xlabel("Standard Deviation")
plt.show()
print(f"From only 100 values the mean is {std100.mean()}")
print(f"From only 100 values the median is {np.median(std100)}")
print(f"The variation from the mean is {np.var(std100)}")
print()
print(f"From 10000 values the mean is {std10000.mean()}")
print(f"From 10000 values the median is {np.median(std10000)}")
print(f"The variation from the mean is {np.var(std10000)}")
```

The more values we use for our normal distribution the closer we get to the true Gaussian distribution with a mean of 3 and standard deviation of 1
