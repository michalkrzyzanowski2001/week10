```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
dataset = pd.read_csv('titanic.csv')
dataset
```

```python
age_values = dataset.loc[dataset['Age'].notnull(), 'Age'].values
print(age_values.shape)
```

```python
plt.hist(age_values, bins=20)
plt.show()
```

```python
mean_age = np.round(np.mean(age_values))
print(mean_age)
plt.hist(age_values, bins=20)
plt.axvline(x = mean_age, color='red', label = 'Mean')
plt.legend(loc='upper right')
plt.show()
```

```python
median_age = np.median(age_values)
print(median_age)
plt.hist(age_values, bins=20)
plt.axvline(x = mean_age, color='red', label = 'Mean')
plt.axvline(x = median_age, color='orange', label = 'Median')
plt.legend(loc='upper right')
plt.show()
```

```python
mode_age = stats.mode(age_values)
print(mode_age)
plt.hist(age_values, bins=20)
plt.axvline(x = mean_age, color='red', label = 'Mean')
plt.axvline(x = median_age, color='orange', label = 'Median')
plt.axvline(x = mode_age[0], color="cyan", label = "Mode")
plt.legend(loc='upper right')
plt.show()
```

```python
q0 = np.quantile(age_values, 0.0)
q1 = np.round(np.quantile(age_values, 0.25))
q3 = np.round(np.quantile(age_values, 0.75))
q4 = np.round(np.quantile(age_values, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_age}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

```python
plt.hist(age_values, bins=20)
plt.axvline(x = mean_age, color='red', label = 'Mean')
plt.axvline(x = median_age, color='orange', label = 'Median')
plt.axvline(x = mode_age[0], color="cyan", label = "Mode")
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.boxplot(age_values)
plt.show()
```

```python
sibsb_values = dataset.loc[dataset['Age'].notnull(), 'SibSp'].values
plt.scatter(x=age_values, y=sibsb_values, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('SibSp')
plt.show()
```

```python
stats.pearsonr(age_values, sibsb_values)
```

```python
fare_values = dataset.loc[dataset['Fare'].notnull(), 'Fare'].values
plt.hist(fare_values, bins=50)
plt.show()
```

```python
mean_fare = np.round(np.mean(fare_values))
median_fare=np.round(np.median(fare_values))
mode_fare=np.round(stats.mode(fare_values))
q0 = np.quantile(fare_values, 0.0)
q1 = np.round(np.quantile(fare_values, 0.25))
q3 = np.round(np.quantile(fare_values, 0.75))
q4 = np.round(np.quantile(fare_values, 1.0))
print(mean_fare)
print(median_fare)
print(mode_fare)
print(q1)
print(q3)
```

```python
plt.figure(figsize=(15,6))
plt.hist(fare_values, bins=102)
plt.axvline(x = mean_fare, color='red', label = 'Mean')
plt.axvline(x = median_fare, color='orange', label = 'Median')
plt.axvline(x = mode_fare[0], color="cyan", label = "Mode")
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
np.max(fare_values) - np.min(fare_values)
```

```python
plt.figure(figsize=(15,3))
plt.boxplot(fare_values,vert=False)
plt.xlabel("Fare Price")
plt.show
```

```python
scaler_standardized = StandardScaler()
standardized_fare_values = scaler_standardized.fit_transform(fare_values.reshape(-1, 1))
plt.hist(standardized_fare_values, bins=20)
plt.axvline(x = standardized_fare_values.mean(), color='red', label = 'Mean')
plt.show()
print(f'Mean: {standardized_fare_values.mean()}')
print(f'Standard Deviation: {np.std(standardized_fare_values, ddof=1)}')
```

```python
scaler_MinMax = MinMaxScaler()
normalized_fare_values = scaler_MinMax.fit_transform(fare_values.reshape(-1, 1))
plt.hist(normalized_fare_values, bins=20)
plt.axvline(x = normalized_fare_values.mean(), color='red', label = 'Mean')
plt.show()
print(f'Mean: {normalized_fare_values.mean()}')
print(f'Standard Deviation: {np.std(normalized_fare_values, ddof=1)}')
```
